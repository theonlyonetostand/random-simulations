Brownian motion from a random walk

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ran_walk_brown( t, s):
    b = np.zeros(np.ceil(t*s) + 1)
    for i in range(np.ciel(t*s) :
        if np.random.uniform(low=0,high=1) > 1/2 :
            b[i+1] = b[i] + np.sqrt(1/s)
        else :
            b[i+1] = b[i] - np.sqrt(1/s)
    x = np.linspace(start=0,stop=t,endpoint=True,num=np.ceil(t*s)+1)
    plt.plot(x,b)
    plt.xlabel('time')
    plt.show()
    return
            
        
```

```python
ran_walk_brown(10,20)
```

```python
def brownian_norm( t, s):
    b = np.zeros(t*s + 1)
    for i in range(t*s):
        b[i+1] = b[i] + np.random.normal(loc=0,scale=1)
    x = np.linspace(start=0,stop=t,endpoint=True,num=t*s+1)
    plt.plot(x,b)
    plt.xlabel('time')
    plt.show()
```

```python
brownian_norm(10,20)
```
